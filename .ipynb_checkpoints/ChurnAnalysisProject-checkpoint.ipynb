**Check for any Outliers** 


```python
# Check for outliers and distribution against target variable Churn and other factors. 
# Are there any factors that suggest they could be directly link to churn or the prevention of churn. 


fig, axes = plt.subplots(3, 3, figsize=(15, 10))
cols = ['Tenure', 'Usage Frequency', 'Support Calls', 'Payment Delay', 
        'Total Spend', 'Last Interaction', 'Age']

for i, col in enumerate(cols):
    r, c = divmod(i, 3)
    sns.boxplot(data=customer_churn, x='Churn', y=col, ax=axes[r,c])
    axes[r,c].set_title(f'{col} by Churn')

plt.tight_layout()
plt.show()
```

****Whats does the Boxplots suggest when comparing each factor against the target variable Churn****
- Payment Delay by Churn: Accounts that have not churned have a lower payment delayed period.
- Total Spend by Churn: Accounts that have a high total spend are less likely to churn, though outliers are present.
- Support Calls by Churn: Accounts that have had less support call (less issues) are less likley to churn, though we are faced with outliers
- Age by Churn: Accounts that have not churned seem to be in the age range of late 20s to mid 40s, worth looking into more 

What percent of Ouliers are there for each variable

```python
num = customer_churn.select_dtypes(include='number')

Q1 = num.quantile(0.25)
Q3 = num.quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = (num < lower) | (num > upper)

outlier_pct = outliers.mean() * 100
print(outlier_pct)
```

**Univariate Analysis**
- churn label indicates whether the customer has churned (1) or not (0)

```python
df = customer_churn
sns.set_style("whitegrid")

# All in one plot
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
for i, col in enumerate(['Age', 'Tenure', 'Total Spend', 'Support Calls', 'Payment Delay', 'Last Interaction']):
    r, c = divmod(i, 3)
    sns.histplot(data=df, x=col, kde=True, ax=axes[r,c], bins=20)
    axes[r,c].set_title(col)

plt.tight_layout()
plt.show()
```

**What does the univariate analysis suggest**


- **Age**: Right skewed/Positively skewed  
- **Total spend**: Left skewed/Negatively skewed
- **Support calls**: Right skewed/Positively skewed  
- **Payment delay**: Right skewed/Positively skewed

```python
corr = customer_churn.corr(numeric_only=True)

sns.heatmap(
    corr,
    annot=True,
    cmap="gray_r",
    center=0,
    linewidths=0.5  # adds thin white lines between cells
)

plt.title('Customer Churn Correlations')  
plt.show()
```

**Observations**

- Churn vs Support calls - Positive strong correlation
The more support calls an customer has are a cause for concern and a moderate indicator of a churn risk
- Churn vs Payment delay - Positive moderate correlation
For accounts with delayed payments is also another moderate indicator for churn
- Churn vs Total spend - Negative moderate correlation
With a negative correlation for Total spend, this indicates that top spenders are less likely to churn


Now that its clear that there are 3 correlating factors that coeside with customers that churn and those that choose to stay.

```python
bins = [18, 25, 35, 45, 55, 65]

labels = ['18-24', '25-34', '35-44', '45-54', '55-64']

customer_churn['Age_Group'] = pd.cut(customer_churn['Age'],
                                    bins=bins,
                                    labels=labels,
                                    right=False)

customer_churn.groupby('Age_Group')['Churn'].mean() * 100
```

```python
data = {
    'Age_Group': ['18-24', '25-34', '35-44', '45-54', '55-64'],
    'Churn_Rate': [55.36, 49.26, 43.46, 52.69, 90.22]
}
df = pd.DataFrame(data)

# Create bar chart
plt.figure(figsize=(10, 6))
plt.bar(df['Age_Group'], df['Churn_Rate'], color='#F57C00', alpha=0.7)

# Customize
plt.title('Churn Rate by Age Group (%)', fontsize=16, fontweight='bold')
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Churn Rate (%)', fontsize=12)
plt.ylim(0, 100)

# Add value labels on top of bars
for i, v in enumerate(df['Churn_Rate']):
    plt.text(i, v + 1, f'{v:.0f}%', ha='center', fontweight='bold')
    
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

***Which subscription type tends to spend more money?*** 



```python
churn_by_age_gender = customer_churn.groupby(['Age_Group', 'Gender'])['Churn'].mean() * 100
print(churn_by_age_gender)
```

```python
data = {
    'Age_Group': ['18-24', '25-34', '35-44', '45-54', '55-64'] * 2,
    'Gender': ['Male']*5 + ['Female']*5,
    'Churn_Rate': [47.4, 41.4, 35.8, 45.0, 89.2,    # Male
                   65.1, 59.3, 53.8, 62.4, 91.1]    # Female
}
df = pd.DataFrame(data)

# Create grouped bar chart (your styling + Male/Female)
plt.figure(figsize=(12, 6))
width = 0.35
x = np.arange(len(df['Age_Group'].unique()[:5]))  # 5 age groups

plt.bar(x - width/2, df[df['Gender']=='Male']['Churn_Rate'].values[:5], 
        width, label='Male', color='#F57C00', alpha=0.7)
plt.bar(x + width/2, df[df['Gender']=='Female']['Churn_Rate'].values[:5], 
        width, label='Female', color='grey', alpha=0.7)

# Your exact customization
plt.title('Churn Rate by Age Group & Gender (%)', fontsize=16, fontweight='bold')
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Churn Rate (%)', fontsize=12)
plt.ylim(0, 100)
plt.xticks(x, ['18-24', '25-34', '35-44', '45-54', '55-64'])
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.3)

# Numbers on bars (your style)
for i, rate in enumerate(df[df['Gender']=='Male']['Churn_Rate'].values[:5]):
    plt.text(i - width/2, rate + 1, f'{rate:.0f}%', ha='center', fontweight='bold')
for i, rate in enumerate(df[df['Gender']=='Female']['Churn_Rate'].values[:5]):
    plt.text(i + width/2, rate + 1, f'{rate:.0f}%', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

**Deeper dive into support calls, what impact does support calls have on churn**
- What is the average amount of calls made by churned1 and not churned0 accounts
- What is the correlation between support calls and churn
- What does the data suggest and how can we implement prevention measures.


```python
variables = ['Contract Length', 'Subscription Type', 'Gender']

for var in variables:
    summary = customer_churn.groupby(var)['Churn'].agg(['count', 'mean']).round(2)
    summary.columns = ['Total_Customers', 'Churn_Rate']
    summary['Churn_Pct'] = summary['Churn_Rate'] * 100
    print(summary)
    print()
```
